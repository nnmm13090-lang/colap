```python
# Creating a stack
def create_stack():
    stack = []
    return stack


# Creating an empty stack
def check_empty(stack):
    return len(stack) == 0


# Adding items into the stack
def push(stack, item):
    stack.append(item)
    print("pushed item: " + item)


# Removing an element from the stack
def pop(stack):
    if (check_empty(stack)):
        return "stack is empty"

    return stack.pop()

stack = create_stack()
push(stack, str(1))
push(stack, str(2))
push(stack, str(3))
push(stack, str(4))
push(stack, str(4))
push(stack, str(5))
push(stack, str(6))
print("popped item: " + pop(stack))
print("stack after popping an element: " + str(stack))
```

```python
import queue
import threading
import time

def demonstate_builting_queues():
  # Demonsstate Pythons build-in queue implenentations

  print("=== PYTHON BUILD-IN QUEUE ===")

# 1. FIFO Queue
  print("\n1. FIFO Queue:")
  fifo_queue = queue.Queue(maxsize=3)

# Add item
  for i in range(3):
    fifo_queue.put(i)
    print(f"Put: {i}")

# Check if full
  print(f"Queue full: {fifo_queue.full()}")

# Get item
  while not fifo_queue.empty():
    item = fifo_queue.get()
    print(f"Got: {item}")

# 2. LIFO Queue (Stack)
  print("\n2. LIFO Queue (Stack): ")
  lifo_queue = queue.LifoQueue()

  for i in range(3):
    lifo_queue.put(i)
    print(f"Pushed: {i}")

  while not lifo_queue.empty():
    item = lifo_queue.get()
    print(f"Poppedd: {item}")
```

```python
# 3. Priority Queue
import queue

print("\n3. Priority Queue:")
priority_queue = queue.PriorityQueue()

items = [(3, 'Low'), (1, 'High'), (2, 'Medium')]
for priority, value in items:
  priority_queue.put((priority, value))
  print(f"Added: {value} (priority: {priority})")

while not priority_queue.empty():
  priority, value = priority_queue.get()
  print(f"Got: {value} (priority: {priority})")
```

```python
class Node:
  def __init__(self, student_id, name, score):
    self.student_id = student_id
    self.name = name
    self.score = score
    self.next = None


class SinglyLinkedList:
  def __init__(self):
    self.head = None

  def append(self, student_id, name, score):
    new_node = Node(student_id, name, score)
    if not self.head:
      self.head = new_node
      return
    cur = self.head
    while cur.next:
      cur = cur.next
    cur.next = new_node

  def delete(self, student_id):
    cur = self.head
    if cur and cur.student_id == student_id:
      self.head = cur.next
      return
    prev = None
    while cur and cur.student_id != student_id:
      prev = cur
      cur = cur.next
    if cur:
      prev.next = cur.next

  def print_list(self):
    cur = self.head
    print("Student ID | Name   | Score")
    print("-" * 30)
    while cur:
      print(f"{cur.student_id:<10} | {cur.name:<8} | {cur.score}")
      cur = cur.next


# Example with real data
students = SinglyLinkedList()
students.append("S01", "Chu", 67)
students.append("S02", "Chi", 68)
students.append("S03", "Tu", 69)
students.append("S04", "Hu", 82)
students.append("S05", "Hi", 89) # Changed S04 to S05 to avoid duplicate IDs

print("Original List:")
students.print_list()

students.delete("S03")
print("\nAfter deleting S03(Charlie)===")
students.print_list()
```

```python
class Node:
  def __init__(self, data):
    self.data = data # The 'data' field
    self.left = None # The 'left' pointer
    self.right = None # The 'right' pointer

def inorder_traversal(root_node):
  """
  Performs a recursive inorder traversal (Left -> Root -> Right)
  of a binary tree and print the data.
  """
  #Base case: if the node is None, do nothing.
  if  root_node:
    # 1. Recursively visit the left subtree
    inorder_traversal(root_node.left)

    # 2. Visit the current Node (print its data)
    # We  use end=" " to print on the same line, just like the C code's space
    print(root_node.data, end=" ")

    # 3. Recursively visite the right subtree
    inorder_traversal(root_node.right)

#----Exsample----
if __name__ == "__main__":

  # Let's build a simple tree manually:
  #       10
  #       / \
  #      5   15
  #     /    /
  #    2     12

  root = Node(10)
  root.left = Node(5)
  root.right = Node(15)
  root.left.left = Node(2)
  root.right.left = Node(12)

  print("Starting Inorder Traversal (left -> Root -> Right):")

  # Call the translated function
  inorder_traversal(root)

  print("\nTraversal conplete.")
```

```python
import csv

SIZE = 50
hash_table = [None] * SIZE

def hash(key):
  return key % SIZE

def insert(key, value):
  index = hash(key)
  while hash_table[index] is not None:
    index = (index + 1) % SIZE
  hash_table[index] = (key, value)

def search(key):
  index = hash(key)
  while hash_table[index] is not None:
    if hash_table[index][0] == key:
      return hash_table[index]
    index = (index + 1) % SIZE
    return None

# Load CSV
with open("/content/drive/MyDrive/students.csv") as f:
  reader = csv.DictReader(f) # Changed to DictReader
  for row in reader:
    insert(int(row["StudentID"]), row["Name"])

# Test search
print(search(1001))
```
